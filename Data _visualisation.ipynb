```python
Ind = pop.iloc[92, 5:13].iloc[::-1]
```

```python
Chi = pop.iloc[41, 5:13].iloc[::-1]
```

```python
def populationdf(country,n):
   
    country = country.reset_index()
    country.rename(columns={n: 'Population'}, inplace=True)
    Year = country['index'].apply(lambda x : x.split(' ')[0])
    country.insert(0,"Year",Year)
    country.drop(['index'],axis=1,inplace=True)
    
    return country
```

```python
Ind = populationdf(Ind,92)
```

```python
Chi = populationdf(Chi, 41)
```

```python
# compare
```

```python
plt.title('Population Growth in India vs China')
plt.ylabel('Popoulation')
plt.xlabel('Year')
plt.plot(Ind['Year'], Ind['Population'], label='India')
plt.plot(Chi['Year'], Chi['Population'], label='China')
plt.legend()
plt.show()
```

```python
# regression line or bestfit line from numpy po;lyfit function
```

```python
fit = np.polyfit(car.carlength, car.carwidth, deg=1)
fit
```

```python
car.plot(kind = 'scatter', x='carlength', y='carwidth', color='g')
plt.title('carlength vs carwidth with the regression line')
plt.plot(car.carlength, fit[0]*car.carlength+fit[1], color='r')
plt.annotate('y={0:.5f}x + {1:.2f}'.format(fit[0], fit[1]),xy=(180,62))
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
alice_novel = open('alice_novel.txt', 'r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
alice_wc = WordCloud(background_color='black', stopwords=stopwords)
alice_wc.generate(alice_novel);
```

```python
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18) 
plt.imshow(alice_wc)
plt.axis('off');
```

```python
stopwords.add('said')
alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18) 

plt.imshow(alice_wc);
```

```python
from PIL import Image
alice_mask = np.array(Image.open('alice.png'))
```

```python
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 

plt.imshow(alice_mask, cmap=plt.cm.gray)
plt.axis('off');
```

```python
alice_wc = WordCloud(background_color='white', stopwords=stopwords, mask=alice_mask)
alice_wc.generate(alice_novel);
fig = plt.figure()
fig.set_figwidth(14)
fig.set_figheight(18) 
plt.imshow(alice_wc)
plt.axis('off');
```
